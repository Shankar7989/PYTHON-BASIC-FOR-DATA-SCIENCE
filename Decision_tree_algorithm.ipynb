```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df.drop(['Id'],axis=1, inplace=True)
```

```python
df['Species']=df['Species'].str.replace('Iris-', '')
```

```python
df.head()
```

```python
df['Species_Num']=df['Species']
```

```python
df.head()
```

```python
df['Species_Num']=df['Species_Num'].str.replace('setosa','1')
df['Species_Num']=df['Species_Num'].str.replace('versicolor','2')
df['Species_Num']=df['Species_Num'].str.replace('virginica','1')
```

```python
df.head()
```

```python
df.drop(['PetalLengthCm','Species_Num'],axis=1,inplace=True)
```

```python
x=df.drop(['Species'],axis=1)
y=df.Species
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(y_test,y_pred)
```

```python
print('a')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
a=cross_val_score(model,x,y,cv=10)
```

```python
from sklearn.metrics import confusion_matrix
a=confusion_matrix(y_test,y_pred)
```
