```python
def fillna_age(cols):
    age=cols[0]
    pclass=cols[1]

    if pd.isnull(age):
        if pclass==1:
            return 38
        elif pclass==2:
             return 30
        else:
             return 23
    else:
         return age
```

```python
df.age = df[['age','pclass']].apply(fillna_age,axis=1)
```

```python
# Assuming your DataFrame is named 'df'

# Define the columns to convert and their mappings
columns_to_convert = {
    'Queenstown': {True: 1, False: 0},
    'Southampton': {True: 1, False: 0},
    'man': {True: 1, False: 0},
    'woman': {True: 1, False: 0},
    'True': {True: 1, False: 0}  # If 'True' is a column name, map it to 1, else to 0
}

# Replace values with 1 or 0
for column, mapping in columns_to_convert.items():
    if column in df.columns:  # Check if column exists in DataFrame
        df[column] = df[column].replace(mapping)
```

```python
from sklearn.model_selection import train_test_split
```
